```python
data['saturated_fat_per_gram'] = (data['saturated_fat_per_gram'] - data['saturated_fat_per_gram'].mean()) / data['saturated_fat_per_gram'].std()
data
```

```python
# finding range by mean + 3 standard deviation so that it covers 99.7% points. points out of range can be considered as outlier
upper_limit = data.saturated_fat_per_gram.mean() + 3*data.saturated_fat_per_gram.std()
lower_limit = data.saturated_fat_per_gram.mean() - 3*data.saturated_fat_per_gram.std()
upper_limit, lower_limit
```

```python
# these are the outliners by saturated fat per gram
outlier_highend_scale = data[(data['saturated_fat_per_gram'] > upper_limit)]  
outlier_lowend_scale  = data[(data['saturated_fat_per_gram'] < lower_limit)]
```

* There are no outliers on lowerend scale

```python
# standardizing all the numerical features

data['weight_in_grams'] = (data['weight_in_grams'] - data['weight_in_grams'].mean()) / data['weight_in_grams'].std()
data['saturated_fat'] = (data['saturated_fat'] - data['saturated_fat'].mean()) / data['saturated_fat'].std()
data['cholesterol'] = (data['cholesterol'] - data['cholesterol'].mean()) / data['cholesterol'].std()

data
```

```python
# finding range by mean + 3 standard deviation so that it covers 99.7% points. points out of range can be considered as outlier
upper_limit_weights = data.weight_in_grams.mean() + 3*data.weight_in_grams.std()
lower_limit_weights = data.weight_in_grams.mean() - 3*data.weight_in_grams.std()
upper_limit_weights, lower_limit_weights
```

```python
# these are the outliners by weight in grams
outlier_weights = data[(data['weight_in_grams'] > upper_limit_weights) | (data['weight_in_grams'] < lower_limit_weights)]
outlier_weights
```

```python
# finding range by mean + 3 standard deviation so that it covers 99.7% points. points out of range can be considered as outlier
upper_limit_cholesterol = data.cholesterol.mean() + 3*data.cholesterol.std()
lower_limit_cholesterol = data.cholesterol.mean() - 3*data.cholesterol.std()
upper_limit_cholesterol, lower_limit_cholesterol
```

```python
# these are the outliners by cholesterol
outlier_cholesterol = data[(data['cholesterol'] > upper_limit_cholesterol) | (data['cholesterol'] < lower_limit_cholesterol)]
outlier_cholesterol
```

* There are some outliers in dataset we can handle that by removing the common outliers from all the features and then use data for further uses.
