```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# importing data
df = pd.read_csv('lifeedata.csv')
```

```python
df
```

# 1.How does Adult Mortality rate affects Life Expectancy ?

```python
df.plot(kind='scatter', x='Adult_Mortality', y='Life_Expectancy')
plt.show()
```

* we can see that as Adult_Mortality rate is increasing as the Life_Expectancy is decreasing.
* therefore as Adult Mortality rate decreases the will be increase in life expectancy of humans.


```python
Pcorr, _ = pearsonr(df['Adult_Mortality'], df['Life_Expectancy'] )
Scorr, _ = spearmanr(df['Adult_Mortality'], df['Life_Expectancy'])
print('Pearson Correlation: ', Pcorr)
print('Spearman Correlation: ', Scorr)
```

```python
covariance = np.cov(df['Adult_Mortality'], df['Life_Expectancy'])
print('Covariance:')
print(covariance)
```

* Both Correlation and Covariance between Adult_mortality rate and human Life_Expectancy is negative.
* Therefore there is negative impact adult mortality on human life expectancy

# Create and plot several linear regression models for life expectancy and independent variables

```python
def LinearRegression(X_train, X_test, y_train, y_test):

    regr = linear_model.LinearRegression()
    
    regr.fit(X_train, y_train)
    
    y_pred = regr.predict(X_test)
    
    print("Mean squared error(MSE): %.2f" % mean_squared_error(y_test, y_pred))
    
    print("Coefficient of determination(R2): %.2f" % r2_score(y_test, y_pred))
    
    plt.scatter(X_test, y_test, color="black")
    plt.plot(X_test, y_pred, color="blue", linewidth=3)
    
    plt.xlabel('Schooling')
    plt.ylabel('Life Expectancy')
    plt.show()
```

## Model of Schooling and Life Expectancy

```python
X = df['Schooling'].values.reshape(-1,1)
y = df['Life_Expectancy'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
LinearRegression(X_train, X_test, y_train, y_test)
```

## Model of Adult Mortality and Life Expectancy

```python
X = df['Adult_Mortality'].values.reshape(-1,1)
y = df['Life_Expectancy'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
LinearRegression(X_train, X_test, y_train, y_test)
```

## Model of BMI and Life Expectancy

```python
X = df['BMI'].values.reshape(-1,1)
y = df['Life_Expectancy'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
LinearRegression(X_train, X_test, y_train, y_test)
```

## Model of Alcohol and Life Expectancy

```python
X = df['Alcohol'].values.reshape(-1,1)
y = df['Life_Expectancy'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
LinearRegression(X_train, X_test, y_train, y_test)
```
